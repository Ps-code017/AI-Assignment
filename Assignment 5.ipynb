```python
import random

items = [('A', 45, 3), ('B', 40, 5), ('C', 50, 8), ('D', 90, 10)]
capacity = 100
population = ['1111', '1000', '0101', '1001']
mutation_order = [3, 2, 1, 0]

def fitness(chromosome):
    weight = value = 0
    for i, gene in enumerate(chromosome):
        if gene == '1':
            weight += items[i][1]
            value += items[i][2]
    return value if weight <= capacity else 0

def select_fittest(pop):
    return sorted(pop, key=fitness, reverse=True)

def crossover(p1, p2):
    return p1[:2] + p2[2:], p2[:2] + p1[2:]

def mutate(chromo, iteration):
    idx = mutation_order[iteration % 4]
    lst = list(chromo)
    lst[idx] = '0' if lst[idx] == '1' else '1'
    return ''.join(lst)

for i in range(10):
    pop_sorted = select_fittest(population)
    new_pop = pop_sorted[:2]
    c1, c2 = crossover(pop_sorted[2], pop_sorted[3])
    c1 = mutate(c1, i)
    new_pop += [c1, c2]
    population = new_pop

for chromo in population:
    print(f"Chromosome: {chromo}, Fitness: {fitness(chromo)}")
```

import math
import random

def evaluate(solution):
    a, b, c, d = map(int, solution)
    clauses = [
        (not a or b),
        (a or (c and b)),
        (not c or not d),
        (a or (not d and not b)),
        (a or not d)
    ]
    return sum(clauses)

def movegen(solution):
    idx = random.randint(0, 3)
    sol = list(solution)
    sol[idx] = '1' if sol[idx] == '0' else '0'
    return ''.join(sol)

current = '1111'
T = 500
cooling = 50
rand_nums = [0.655, 0.254, 0.432]
i = 0

while T > 0:
    next_sol = movegen(current)
    delta = evaluate(next_sol) - evaluate(current)
    if delta > 0:
        current = next_sol
    else:
        prob = math.exp(delta / T)
        if rand_nums[i % len(rand_nums)] < prob:
            current = next_sol
    T -= cooling
    i += 1

print(f"Final Solution: {current}, Clauses Satisfied: {evaluate(current)}")
